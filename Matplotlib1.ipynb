```python
import matplotlib.pyplot as plt
# data to display on plots
x = [3, 1, 3]
y = [3, 2, 1]

# This will plot a simple line chart
# with elements of x as x axis and y
# as y axis
plt.plot(x, y)
plt.title("Line Chart")

# Adding the legends
plt.legend(["Line"])
plt.show()
```

```python
plt.plot(x, y)
plt.title("Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Adding the legends
plt.legend(["Line"])
plt.show()
```

```python
import numpy as np

x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))

plt.stem(x, y, use_line_collection = False)
plt.show()
```

```python
help(plt.stem)
```

```python
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]

# This will plot a simple bar chart
plt.bar(x, y)

# Title to the plot
plt.title("Bar Chart")

# Adding the legends
plt.legend(["bar"],labelcolor="red")
plt.show()
plt.bar(x,y,linestyle="--",color=["green","red","orange","blue"])
plt.show()
```

```python
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]

# This will plot a simple scatter chart
plt.scatter(x, y,marker='o',color="green")

# Adding legend to the plot
plt.legend("abc")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Title to the plot
plt.title("Scatter chart")
plt.show()
```

```python
help(plt.scatter)
```

```python
days = [1, 2, 3, 4, 5]

# No of Study Hours
Studying = [7, 8, 6, 11, 7]

# No of Playing Hours
playing = [8, 5, 7, 8, 13]

# Stackplot with X, Y, colors value
plt.stackplot(days, Studying, playing,
            colors =['r', 'c'])

# Days
plt.xlabel('Days',color="blue",fontname="Arial",fontsize=10,fontweight='bold')

# No of hours
plt.ylabel('No of Hours',color="blue",fontname="Times New Roman")

# Title of Graph
plt.title('Representation of Study and \
Playing wrt to Days')
plt.legend(['Studying', 'Playing'])

# Displaying Graph
plt.show()
```

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

# Creating plot
plt.boxplot(data,patch_artist=True,boxprops=dict(facecolor='orange'))
# show plot
plt.show()
```

```python
help(plt.boxplot)
```

```python
x = [1, 2, 3, 4]

# this will explode the 1st wedge
# i.e. will separate the 1st wedge
# from the chart
e  =(0.1, 0.1, 0.1, 0.1)
labels=['a','b','c','d']
# This will plot a simple pie chart
plt.pie(x, center=(-1,-1),labels=labels,explode = e)
plt.legend(["pie"],labelcolor="red")

# Title to the plot
plt.title("Pie chart")
plt.show()
```
